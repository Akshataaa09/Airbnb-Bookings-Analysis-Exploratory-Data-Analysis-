## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# Checking missimg values
df.isnull().sum()
````

````python
# Handling missing values
df.dropna(inplace = True)
df.shape
````

````python
# Dropping unnecessary columns
df.drop(['host_name'],axis= 1,inplace= True)
````

````python
# Type Conversion
# Convert 'last_review' to datetime
df['last_review'] = pd.to_datetime(df['last_review'], errors='coerce')
````

````python
# Checking for outliers
sns.boxplot(x =df['price'])
plt.title("Outlier Detection - Price")
plt.show()
````

````python
# Removing outliers

# Assuming df is already defined
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
IQR = Q3 - Q1

# Define the bounds for non-outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter out the outliers
df_no_outliers = df[(df['price'] >= lower_bound) & (df['price'] <= upper_bound)]
````

````python
# New boxplot after removing outliers
sns.boxplot(x=df_no_outliers['price'])
plt.title("Price After Removing Outliers")
plt.show()
````

````python
# Handling invalid data
# Check for zero or negative values in 'minimum_nights'
df = df[df['minimum_nights'] > 0]
````

````python
# Check for latitude and longitude bounds
# Define expected coordinate boundaries for NYC
lat_min, lat_max = 40.5, 40.92
long_min, long_max = -74.25, -73.70

# Find listings outside these bounds
invalid_coords = df[
    (df['latitude'] < lat_min) | (df['latitude'] > lat_max) |
    (df['longitude'] < long_min) | (df['longitude'] > long_max)
]

# Show how many invalid coordinates are found
print("Invalid Coordinate Entries:", invalid_coords.shape[0])

# Display a few invalid rows (if any)
invalid_coords[['id', 'latitude', 'longitude', 'name']].head()

````

````python
# Feature engineering
# Extracting year and month from Date
df['last_review_year'] = df['last_review'].dt.year

df['last_review_year']
````

````python
# Average Price per Neighbourhood Group
avg_price = df.groupby('neighbourhood_group')['price'].mean().round(2).sort_values(ascending = False)
print(avg_price)
````

````python
# Listing per Room Type
room = df.groupby('room_type')['price'].mean().round(2)
print(room)
````

# Statistical Analysis

**1. Descriptive statistics for 'price'**

````python
# Basic descriptive statistics for 'price'
from scipy import stats

price_data = df['price']
price_data = price_data[price_data < 500]  # Removing extreme outliers

print(f"Minimum Price: {price_data.min()}")
print(f"Maximum Price: {price_data.max()}")
print(f"Mean Price: {price_data.mean():.2f}")
print(f"Median Price: {price_data.median()}")
print(f"Mode Price: {price_data.mode().iloc[0]}")  # mode() can return multiple values
````

**2. Hypothesis Testing using t-test**: Price Difference Between Manhattan and Brooklyn

Hypothesis:
H0: There is no significant difference in average price between listings in Manhattan and Brooklyn

H1: There is a significant difference

````python
from scipy.stats import ttest_ind

# Filter dataset for only Manhattan and Brooklyn
manhattan = df[df['neighbourhood_group'] == 'Manhattan']['price']
brooklyn = df[df['neighbourhood_group'] == 'Brooklyn']['price']

# Remove outliers for fairness (optional)
manhattan = manhattan[manhattan < 500]
brooklyn = brooklyn[brooklyn < 500]

# Perform independent t-test
t_stat, p_value = ttest_ind(manhattan, brooklyn, equal_var=False)

print(f"T-statistic: {t_stat:.2f}")
print(f"P-value: {p_value:.4f}")

if p_value < 0.05:
    print("Conclusion: Reject Null Hypothesis – There is a significant difference in prices.")
else:
    print("Conclusion: Fail to Reject Null – No significant difference in prices.")
````
